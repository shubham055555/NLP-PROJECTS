```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

dataset = "/content/Copy of Synthetic News Dataset - Synthetic News Dataset.csv"

# Combine text
text = " ".join(dataset)

# Remove stopwords
stop_words = set(stopwords.words("english"))
words = [w.lower() for w in text.split() if w.lower() not in stop_words]
clean_text = " ".join(words)

# Generate word cloud
wc = WordCloud(width=1000, height=500, background_color="white").generate(clean_text)

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
