```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
context = """
Machine learning is a field of artificial intelligence that uses statistical
techniques to give computer systems the ability to learn from data.
Supervised learning uses labeled data, while unsupervised learning uses
unlabeled data to find patterns. Neural networks are powerful models inspired
by the human brain.
"""
```

```python
def ask_bot(question):
    result = qa_pipeline(question=question, context=context)
    return result["answer"]
```

```python
print(ask_bot("What is machine learning?"))
print(ask_bot("What is supervised learning?"))
print(ask_bot("What are neural networks?"))
```
