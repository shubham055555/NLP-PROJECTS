```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

print("âœ… Libraries Imported Successfully!\n")
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

print("âœ… Pretrained Sentence-BERT Model Loaded!\n")
```

```python
sentences = [
    "I love playing football with my friends.",
    "Soccer is my favorite sport.",
    "The sun is shining bright today.",
    "Artificial Intelligence is transforming the world.",
    "I enjoy watching movies on weekends."
]

print("âœ… Sentences Loaded:")
for s in sentences:
    print("-", s)
print("\n")
```

```python
embeddings = model.encode(sentences)

print("âœ… Sentence Embeddings Generated! Shape:", embeddings.shape, "\n")
```

```python
cosine_sim = cosine_similarity(embeddings)

print("ðŸ“Š Cosine Similarity Matrix:")
print(np.round(cosine_sim, 3), "\n")
```

```python
query = "I like to play soccer."
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]
most_similar_idx = np.argmax(similarities)

print(f"ðŸ” Query: {query}")
print(f"Most Similar Sentence: {sentences[most_similar_idx]}")
print(f"Similarity Score: {similarities[most_similar_idx]:.3f}\n")
```

```python
corpus = [
    "The sky is clear and blue.",
    "I am going to play football after lunch.",
    "Artificial Intelligence and Machine Learning are related fields.",
    "It might rain today, take an umbrella!",
    "He scored a goal in the last match."
]

queries = [
    "What is AI?",
    "Who won the football match?",
    "Will it rain today?"
]

# Encode corpus and queries
corpus_embeddings = model.encode(corpus, convert_to_tensor=True)
query_embeddings = model.encode(queries, convert_to_tensor=True)

print("âœ… Corpus and Queries Encoded!\n")

# Perform semantic search
for i, query in enumerate(queries):
    scores = util.cos_sim(query_embeddings[i], corpus_embeddings)[0]
    best_match = np.argmax(scores)
    print(f"ðŸ”¹ Query: {query}")
    print(f"Best Match: {corpus[best_match]}")
    print(f"Similarity Score: {scores[best_match]:.3f}\n")
```
